## 5.Используя pandas, проведи когортный анализ пользователей. В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц.

```python
# Формируем DataFrame
users_order_5 = users[['customer_unique_id', 'customer_id']]\
                    .merge(order[order['order_status'] == "delivered"][['customer_id', 'order_approved_at']],\
                           how = 'inner', on = 'customer_id').copy()
```

```python
#Извлекаем месяца 
users_order_5['month'] = users_order_5['order_approved_at'].dt.to_period('M')
```

```python
users_order_5.head()
```

```python
#Ищем дату первого заказа пользователя
first = users_order_5.sort_values(by='month')\
    .groupby('customer_unique_id').agg({'month': 'first'}).rename(columns={'month':'first_month'})
```

```python
users_order_5 = users_order_5.merge(first,how = 'inner', on = 'customer_unique_id').drop('order_approved_at', axis = 1)
```

```python
users_order_5.head()
```

```python
users_order_5.shape
```

```python
users_order_5 = users_order_5.groupby(['first_month', 'month'], as_index=False).agg({'customer_unique_id':'count'})
```

```python
#Выведем столбец с покупками в первый месяц когорты
hz = users_order_5.groupby('first_month').agg({'customer_unique_id':'max'}).copy()
```

```python
hz.T
```

```python
users_order_5.head()
```

```python
#Преобразуем в нужный формат
users_order_rate = pd.pivot(users_order_5, index='first_month', columns='month', values='customer_unique_id')
```

```python
users_order_rate.head()
```

```python
#Посчитаем retention 
users_order_rate = (users_order_rate.T / hz['customer_unique_id']).T
```

```python
users_order_rate
```

```python
plt.figure(figsize=(18,18))
sns.heatmap(round(users_order_rate, 4), annot = True, vmin=0.000001, vmax=0.01, cmap= 'coolwarm', linewidths=1, linecolor='black',\
           cbar_kws={'orientation': 'horizontal'})
```

```python
#Определяем когорту у которой retention больше всего на 3 месяц
pr = users_order_rate.fillna(0).copy()
k = pr.shape[0]
mx = 0
for i in range(0,k-3):
    if pr.iat[i, i+2] > mx:
        mx = pr.iat[i, i+3]
        name = pr.index[i]
print('Когорта:  ', name, '\n' + 'retention:', mx.round(5))
```

## 6.Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты, F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. 

```python
# Формируем DataFrame
users_order_product_6 = users[['customer_unique_id', 'customer_id']]\
                    .merge(order[(order['order_status'] == "delivered")\
                                  & (order['order_approved_at'].isna() == False)][['customer_id', 'order_id']],\
                           how = 'inner', on = 'customer_id').copy()
users_order_product_6 = users_order_product_6.merge(product_order[['order_id', 'price']], how = 'inner', on = 'order_id').copy()
```

```python
users_order_product_6 = users_order_product_6.sort_values(by = 'customer_unique_id')
```

```python
users_order_product_6.head()
```

Так как в чеке не указано кол-во товаров, то общую сумму по чеку посчитаем просто сложением цены всех товарных позиций указанных в нём и примёрджим дату покупки

```python
users_order_product_6 = users_order_product_6.groupby(['customer_unique_id', 'customer_id'], as_index=False)\
                        .agg({'price' : 'sum'})
```

```python
users_order_product_6.head()
```

```python
#Так как у нас уже есть сумма по чеку колонка order_id не нужна
users_order_product_6 = users_order_product_6.merge(order[['customer_id', 'order_approved_at']],\
                                    how = 'inner', on = 'customer_id').copy()
```

```python
users_order_product_6 = users_order_product_6[['customer_unique_id', 'customer_id', 'order_approved_at', 'price']]
```

```python
users_order_product_6.head()
```

```python
users_order_product_6.dtypes
```

Датафрейм подготовили теперь расчитаем Recency, Freqency и Monetary

```python
from datetime import timedelta
```

```python
#Возьмём точку отсчёта времени, +1 день к самой ранней дате
last_day = users_order_product_6['order_approved_at'].max() + timedelta(days=1)
last_day
```

```python
#Создаём столбец с кол-вом дней от совершения покупки
users_order_product_6['Days'] = (last_day - users_order_product_6['order_approved_at']).dt.days.astype('Int32')
```

```python
users_order_product_6
```

```python
#Расчитываем Recency, Frequency и Monetary
users_order_product_6 = users_order_product_6.groupby('customer_unique_id', as_index=False)\
            .agg({'Days' : 'min', 'order_approved_at':'count', 'price' : 'sum'})\
            .rename(columns={'Days': 'Recency', 'order_approved_at': 'Frequency', 'price':'Monetary'})
```

```python
users_order_product_6.head()
```

```python
#Расчитаем квантили, будем делить каждую метрику на 3 группы
quintiles = users_order_product_6[['Recency', 'Frequency', 'Monetary']].quantile([.33, .66]).to_dict()
quintiles
```

```python
(users_order_product_6[users_order_product_6['Frequency'] > 1].shape[0]*100 / users_order_product_6.shape[0])
```

```python
def r_score(x):
    if x <= quintiles['Recency'][.33]:
        return 3
    elif x <= quintiles['Recency'][.66]:
        return 2
    else:
        return 1
```

```python
def m_score(x, c):
    if x <= quintiles[c][.33]:
        return 1
    elif x <= quintiles[c][.66]:
        return 2
    else:
        return 3   
```

```python
{'Recency': {0.33: 147, 0.66: 286},
 'Frequency': {0.33: 1.0, 0.66: 1.0},
 'Monetary': {0.33: 57.9, 0.66: 124.9}}
```

Разобьём пользователей на 3 категории, так чтобы примерно одно и тоже количество попало в категорию.
Как видим фирма специализируется на разовых продажах и клиентов, которые делают более одной покупки всего 3%, поэтому для Frequency будет две категории. В итоге у нас должно получиться 18 кластеров пользователей

Recency: (0-147 days) - 3,  (147-286 days) - 2, (286 days <) - 1;

Frequency: 1 - 1, 1< - 2;
        
Monetary: (0-57.9$) - 1,  (57.9-124.9$) - 2, (124.9<) - 3;

```python
users_order_product_6['R'] = users_order_product_6['Recency'].apply(lambda x: r_score(x))
users_order_product_6['F'] = users_order_product_6['Frequency'].apply(lambda x: 1 if x == 1 else 2)
users_order_product_6['M'] = users_order_product_6['Monetary'].apply(lambda x: m_score(x, 'Monetary'))
```

```python
users_order_product_6['RFM_Score'] = (users_order_product_6['R'].map(str)\
                                    + users_order_product_6['F'].map(str)\
                                    + users_order_product_6['M'].map(str)).astype('Int32')
```

```python
users_order_product_6 = users_order_product_6.sort_values(by = 'RFM_Score', ascending=False)
```

```python
users_order_product_6
```

```python
hz = pd.DataFrame(users_order_product_6['RFM_Score'].value_counts()).reset_index()\
            .rename(columns={'index' : 'RFM_Score', 'RFM_Score':'Count'})
```

```python
plt.figure(figsize=(18,6))
plt.title('График пользователей в кластерах', fontsize= 25)
plt.tick_params (axis='both', which='major', labelsize= 15 )
sns.barplot(data = hz , x='RFM_Score', y= 'Count')
plt.ylabel('')
plt.xlabel('RFM оценка',fontsize= 15)
```

Для того, чтобы пользователю было проще воспринимать оценку, для работы с кластерами выразим её не в цифрах, а в характеристиках. Для этого добавим группы классификации, основанные на RFM оценки.

```python
def classify(x):
    k = ''
    if x // 100 == 1:
        k += 'Потерянные, '
    elif x // 100 == 2:
        k += 'Спящие, '
    else:
        k += 'Новые, '
    if (x % 100) // 10 == 1:
        k +='разовые и '
    else:
        k +='постоянные и '
    if x % 10 == 1:
        k += 'с маленьким чеком'
    elif x % 10 == 2:
        k += 'со средним чеком'
    else:
        k += 'с большим чеком'
        
    return k 
users_order_product_6['Сlassify'] = users_order_product_6['RFM_Score'].apply(classify)
```

```python
users_order_product_6
```

```python
hz = pd.DataFrame(users_order_product_6['Сlassify'].value_counts()).reset_index()\
            .rename(columns={'index' : 'Сlassify', 'Сlassify':'Count'})
```

```python
hz
```

```python
plt.figure(figsize=(18,12))
plt.title('График классификации пользователей', fontsize= 30)
#plt.yticks(rotation=0, ha='right')
sns.barplot(data = hz , x='Count', y= 'Сlassify')
plt.xlabel('Количество пользователей', fontsize= 20)
plt.ylabel('') 
plt.tick_params (axis='both', which='major', labelsize= 20 ) 
```
